##**Insert and Delete Columns and Replace column values**

```python
df1=df.copy()
df1.head()
```

```python
df=df1[['INVOICENO','UNITPRICE','CUSTOMERID','COUNTRY']]
df.head()
```

```python
df1.drop('INVOICENO', axis = 1, inplace = True)  # deleting column INVOICENO
df1.head()
```

```python
df1.insert(1, "TAX", df1.UNITPRICE * 0.10)
df1.head()
```

```python
df1.COUNTRY.value_counts()
```

```python
df1.COUNTRY.replace({'France':'FR','United Kingdom':'USA','Australia':'AUS','Netherlands':'NED'},inplace=True)
df1.COUNTRY.value_counts()
df1.head()
```

```python
df1['COUNTRY'].replace(["Fr","UK","AUS","NED","GER"],["France","United kingdom",'Australia','Netherlands','Germany'],inplace=True)
df1
```

```python
print(pd.factorize(df1['COUNTRY'])[0])
df1.sort_values(by='UNITPRICE',ascending=True,ignore_index=True).head()
```

```python
#Factorizing country column Values
df1.insert(4,'COUNTRY_CODE',pd.factorize(df1['COUNTRY'])[0])
print(pd.factorize(df1['COUNTRY'])[0])
df1.head()
```
