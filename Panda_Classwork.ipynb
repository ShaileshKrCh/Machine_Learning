# **Introduction to Pandas**
It is often said that 80% of data analysis activity is spent on the data cleaning and preparation of our data. To get the grip over this activity, this section will focus on a small but important aspect of data manipulation and cleaning with Pandas.

**Inbuilt Data Structure in Pandas**

There are two types of data structures are there in Pandas -

**Series** - It is one-dimensional labeled array capable of holding any data type (integer, string, float, Python objects etc.) of data. The axis is collectively referred to as index.

**Data Frame** - It is a two-dimensional labeled data structure with columns of potentially different types. You can think of it like a spreadsheet or SQL Table or a Series of objects

**Series Data Structure**
pandas.core.series.Series(data, index, dtype, copy)
* data - data takes various forms like ndarray, list, constants, dictionary etc.
* index - it is unique and hashable for easy identification.
* dtype - it is for data type.
* copy - copy data, and its default value is False. It only affects for Series or one dimensional ndarray inputs.

**Class Assignment:**

input words: ['alaSKa', 'mango', 'dAd', 'coLLege']

output words: ['alaSKa', 'dAd']
All the characters must remain present in a single row of the keyboard.

```python
myword ="univerSIty"
print(myword,set(myword.lower()))
```

```python
# Input list
words = ['alaSKa', 'mango', 'dAd','sdfg','coLLege']

# Define the rows of the keyboard
row1 = set('qwertyuiop')
row2 = set('asdfghjkl')
row3 = set('zxcvbnm')

# Filter the words
result = []
for word in words:
    lower_word = set(word.lower())
    if lower_word.issubset(row1) or lower_word <= row2 or lower_word <= row3:
        result.append(word)
print(result)
```

# Join Operations on DataFrames

download (1).png

```python
data_dict = dict(name=['alice','john','eric','julie','tom'], height=[165,180,175,180,185])
print(data_dict)
user4 = pd.DataFrame(data = data_dict)
user4
```

```python
# inner join: All matching rows from both dataframes
merge_inner = pd.merge(users, user4, on='name')  # by default inner join
merge_inner
```

```python
merge_inner = pd.merge(users, user4, on='name', how='inner')
merge_inner
```

```python
# left outer join: All rows from first and matching rows from second dataframes
merge_left = pd.merge(users, user4, on='name', how='left')
merge_left
```

```python
# right outer join: All rows from second and matching rows from first dataframes
merge_right = pd.merge(users, user4, on='name', how='right')
merge_right
```

```python
# outer join: All rows matching and not matching from both dataframes
merge_outer = pd.merge(users, user4, on='name', how='outer')
merge_outer
```

# **Pivot and Unpivot operations**

**Unpivot Operation:**

Converts from Wide Format (Unstacked Format) ----to----> Long Format (Stacked Format)

**Pivot Operation:**

 Converts from Long Format (Stacked Format) ----to----> Wide Format (Unstacked Format)

```python
df= users.copy()
df
```

```python
# unpivot operation
stacked = pd.melt(df, id_vars="name", var_name='variable', value_name='value')  # long / stacked format
stacked
```

```python
stacked.info()
```

```python
# pivot operation
unstacked = stacked.pivot(index="name", columns="variable", values="value")  # wide / unstacked format
unstacked.reset_index(inplace=True)
unstacked.columns.name = ""
unstacked
```
